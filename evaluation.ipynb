```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Flatten

class TrendAnalyzer:
    def __init__(self, model_path='trading_trend_model.h5', img_size=(64, 64)):  # R√©duire la taille
        self.model = load_model(model_path)
        self.img_size = img_size
        
        # V√©rification de la compatibilit√© du mod√®le
        self.check_model_compatibility()
        
    def check_model_compatibility(self):
        """V√©rifie que les dimensions du mod√®le correspondent √† nos entr√©es"""
        sample_input = np.zeros((1, 10, *self.img_size, 1))
        try:
            self.model.predict(sample_input)
        except Exception as e:
            print("\n‚ö†Ô∏è Erreur de compatibilit√© du mod√®le:")
            print(str(e))
            print("\nüîß Solution recommand√©e:")
            print("1. V√©rifiez que la taille d'image (img_size) correspond √† celle utilis√©e pendant l'entra√Ænement")
            print("2. R√©entra√Ænez le mod√®le avec les bonnes dimensions si n√©cessaire")
            print(f"3. Taille d'image actuelle: {self.img_size}")
            exit()
    
    def preprocess_image(self, image_path):
        """Pr√©traitement d'une image pour le mod√®le"""
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"‚ö†Ô∏è Image non lue: {image_path}")
            return None
            
        day_width = img.shape[1] // 10
        sequence = []
        
        for i in range(10):
            day_img = img[:, i*day_width:(i+1)*day_width]
            day_img = cv2.resize(day_img, self.img_size)
            sequence.append(day_img)
        
        return np.array(sequence).reshape(1, 10, *self.img_size, 1) / 255.0
    
    def predict_image(self, image_path):
        """Pr√©diction pour une seule image"""
        processed_img = self.preprocess_image(image_path)
        if processed_img is None:
            return None
            
        try:
            pred = self.model.predict(processed_img)[0][0]
            trend = "Uptrend" if pred > 0.5 else "Downtrend"
            
            # Calcul des scores par jour
            day_scores = []
            for i in range(10):
                day_pred = self.model.predict(processed_img[:, i:i+1])[0][0]
                day_scores.append(float(day_pred))
                
            return {
                'trend': trend,
                'confidence': max(pred, 1 - pred),
                'raw_score': float(pred),
                'day_scores': day_scores
            }
        except Exception as e:
            print(f"Erreur de pr√©diction sur {os.path.basename(image_path)}: {str(e)}")
            return None
    
    def analyze_folder(self, folder_path, true_label):
        """Analyse toutes les images d'un dossier"""
        results = []
        valid_images = 0
        
        for img_file in sorted(os.listdir(folder_path)):
            if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
                img_path = os.path.join(folder_path, img_file)
                pred = self.predict_image(img_path)
                
                if pred is not None:
                    results.append({
                        'file': img_file,
                        'true_label': true_label,
                        'predicted': pred['trend'],
                        'confidence': pred['confidence'],
                        'correct': (true_label.lower() == pred['trend'].lower())
                    })
                    valid_images += 1
                    self.visualize_single(img_path, pred, true_label)
        
        print(f"\nüîπ {valid_images} images valides analys√©es dans {folder_path}")
        return results
    
    def visualize_single(self, img_path, prediction, true_label):
        """Visualisation pour une image"""
        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        
        plt.figure(figsize=(15, 8))
        
        # Image
        plt.subplot(2, 1, 1)
        plt.imshow(img)
        plt.title(f"{os.path.basename(img_path)}\nVrai: {true_label} | Pr√©dit: {prediction['trend']} ({prediction['confidence']:.2%})")
        plt.axis('off')
        
        # Courbe des jours
        plt.subplot(2, 1, 2)
        plt.plot(range(1, 11), prediction['day_scores'], 'b-o')
        plt.axhline(0.5, color='r', linestyle='--')
        plt.xlabel('Jour')
        plt.ylabel('Score Uptrend')
        plt.xticks(range(1, 11))
        plt.ylim(0, 1)
        plt.grid(True)
        
        plt.tight_layout()
        plt.show()
    
    def generate_report(self, uptrend_results, downtrend_results):
        """G√©n√®re un rapport global"""
        all_results = uptrend_results + downtrend_results
        
        if not all_results:
            print("\n‚ö†Ô∏è Aucune image valide analys√©e!")
            print("V√©rifiez que:")
            print("- Les images existent dans les dossiers")
            print("- Leurs dimensions sont compatibles avec le mod√®le")
            print(f"- Taille d'image attendue: {self.img_size}")
            return
        
        accuracy = sum(r['correct'] for r in all_results) / len(all_results)
        
        print("\n" + "="*50)
        print(f"üìä RAPPORT COMPLET (Total: {len(all_results)} images valides)")
        print(f"üîπ Accuracy globale: {accuracy:.2%}")
        print("="*50)
        
        # Matrice de confusion
        true_pos = sum(1 for r in all_results if r['true_label'] == 'Uptrend' and r['predicted'] == 'Uptrend')
        false_pos = sum(1 for r in all_results if r['true_label'] == 'Downtrend' and r['predicted'] == 'Uptrend')
        true_neg = sum(1 for r in all_results if r['true_label'] == 'Downtrend' and r['predicted'] == 'Downtrend')
        false_neg = sum(1 for r in all_results if r['true_label'] == 'Uptrend' and r['predicted'] == 'Downtrend')
        
        cm = [[true_pos, false_neg], [false_pos, true_neg]]
        
        plt.figure(figsize=(8, 6))
        plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
        plt.title("Matrice de confusion")
        plt.colorbar()
        plt.xticks([0, 1], ['Pr√©dit Uptrend', 'Pr√©dit Downtrend'])
        plt.yticks([0, 1], ['Vrai Uptrend', 'Vrai Downtrend'])
        
        for i in range(2):
            for j in range(2):
                plt.text(j, i, str(cm[i][j]), ha="center", va="center", 
                        color="white" if cm[i][j] > max(cm[0][0], cm[1][1])/2 else "black")
        
        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    # Configuration
    dataset_dir = "30_dataset"
    uptrend_dir = os.path.join(dataset_dir, "uptrend")
    downtrend_dir = os.path.join(dataset_dir, "downtrend")
    
    # V√©rification des dossiers
    if not os.path.exists(uptrend_dir) or not os.path.exists(downtrend_dir):
        print(f"Erreur: Structure de dossier incorrecte")
        print("Structure attendue:")
        print("30_dataset/")
        print("‚îú‚îÄ‚îÄ uptrend/  (contient les images Uptrend)")
        print("‚îî‚îÄ‚îÄ downtrend/  (contient les images Downtrend)")
        exit()
    
    print("üîç D√©marrage de l'analyse...")
    analyzer = TrendAnalyzer(img_size=(64, 64))  # Ajustez selon votre mod√®le
    
    # Analyse
    print("\nüîç Analyse des Uptrends...")
    uptrend_results = analyzer.analyze_folder(uptrend_dir, "Uptrend")
    
    print("\nüîç Analyse des Downtrends...")
    downtrend_results = analyzer.analyze_folder(downtrend_dir, "Downtrend")
    
    # Rapport final
    analyzer.generate_report(uptrend_results, downtrend_results)
    print("\n‚úÖ Analyse termin√©e!")
```
